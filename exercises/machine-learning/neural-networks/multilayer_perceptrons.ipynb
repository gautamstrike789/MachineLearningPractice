# Importing Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
import sklearn.metrics
```

# Loading the Dataset

```python
df = pd.read_csv("/content/mnist.csv")
df.set_index('id', inplace=True)
df.head(5)
```

# Splitting Data into Train and Test

```python
x = df.drop(["class"], axis=1)
y = df["class"]
x_train, x_test, y_train, y_test = train_test_split(x, y)
print("x_train.shape:", x_train.shape)
print("x_test.shape:", x_test.shape)
print("y_train.shape:", y_train.shape)
print("y_test.shape:", y_test.shape)
```

# Algorithm Selection and Hyperparameter Tuining

```python
parameters_grid = {
    "solver" : ["lbfgs", "sgd", "adam"],
    "batch_size" : [10, 20, 50, 100, 200],
    "max_iter" :[1000, 2000],
    "hidden_layer_sizes": [(50), (100), (50, 50), (50, 100), (100, 50), (100, 100)]
}
model_1 = GridSearchCV(MLPClassifier(), parameters_grid, scoring="accuracy", cv=5, n_jobs=-1)
model_1.fit(x_train, y_train)
print("Accuracy of best MLP classifier = {:.2f}".format(model_1.best_score_))
print("Best parameters of best MLP classifier = {}".format(model_1.best_params_))
```
